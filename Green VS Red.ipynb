Green VS Red

```python
import pandas as pd
import numpy as np
import random as random
from matplotlib import pyplot as plt
from matplotlib import colors


class Green_VS_Red():
    def __init__(self):
        self.number_of_generations = 0
        self.number_of_times_chosen_cell_is_green = 0
    
    def initialise_grid(self, x, y):
        if (x < 1) | (y < 1) | (x > y) | (y > 1000):
            # If the grid dimensions are out of the required values then the user is notfied and prompted to re-enter the arguments.
            # Sets the grid attribute to None which haults progression and processessing by subsequent methods.
            print('Invalid arguments')
            print('\nPlease enter an x and y value where x <= y <= 1000')
            self.grid = None
        else:
            # If valid grid dimensions are met a dataframe is created with the specific width (x) and height (y) defined by the user.
            # The grid is then initialised for Generation Zero, populating the grid randomly with 1s (green) and 0s (red) by performing a dataframe concat for each row (y) of width (x). 
            # The dataframe has column and index numbers prefixed with their respective x or y label. 
            self.x_range = x
            self.y_range = y
            self.grid = pd.DataFrame(np.array([[random.choice([1, 0]) for i in range(x)]]), index=['y'+str(1)], columns = ['x'+str(i+1) for i in range(x)] )
            for i in range(y-1):
                new_line = pd.DataFrame(np.array([[random.choice([1, 0]) for i in range(x)]]), index = ['y'+str(i+2)], columns=  ['x'+str(j+1) for j in range(x)])
                self.grid = pd.concat([self.grid, new_line])
            if self.grid['x'+str(self.chosen_x)].loc['y'+str(self.chosen_y)] == 1:
                # This includes the initial board state (Generation Zero) in the number of times the chosen cell is green.
                self.number_of_times_chosen_cell_is_green += 1 
            else:
                # If the chosen cell is initially red (0) then the counter stays at nil.
                pass
            print('Initialising grid..')
            print('Generation Zero')
            print(f'So far your coordinates ({self.chosen_x},{self.chosen_y}) were green {self.number_of_times_chosen_cell_is_green} time(s)')
            # The dataframe of the grid is displayed in the console
            display(self.grid)
    
    def visualise_grid(self):
        if isinstance(self.grid, pd.DataFrame):
            # If the grid variable is of type DataFrame then plot the grid using matplotlib and colour the board according to its value.
            # The figure size of the board is dynamically adjusted with respect to the grid dimensions through the scaling factor variables. 
            # The plot layout is configured to label the board coordinates allowing the left and top of the plot.
            # The x and y ticks are configured to occur every increment in coordinate position.
            # The x and y tick labels sit at the centre of each position by adjusting the loc argument of the plt.(x|y)ticks() method by 0.5  
            cmap = colors.ListedColormap(['red','green'])
            scaling_factor_x = self.x_range/20 + 0.5
            scaling_factor_y = self.y_range/20 + 0.5
            fig, ax = plt.subplots(figsize=(6*scaling_factor_x, 6*scaling_factor_y))
            plt.pcolor(self.grid[::-1],cmap=cmap,edgecolors='k', linewidths=3)
            plt.xticks([i+0.5 for i in range(self.x_range)], [i+1 for i in range(self.x_range)])
            plt.yticks([i+0.5 for i in range(self.y_range)], [i+1 for i in range(self.y_range)[::-1]])
            ax.xaxis.tick_top()
            fig.tight_layout()
            self.plot = plt.show()
            return self.plot
        else:
            # If the grid variable is not of type DataFrame (in the instance where the grid is improperly defined) then this method does nothing.
            return
    
    def rules_for_colour_change(self, x_coordinate, y_coordinate):
        # Here variables are defined for the values surrounding the current value.
        # The rules on how the colour changes are then applied to COPY of the dataframe.
        # This copy is the new state that will exist once all the rules are applied.
        # This eliminates the bias towards the top left of the dataframe as the processing order inherently changes values which would propegate through to values later in the process order.
        # This ensures all rules are applied sumultaniously to generate the next board state generation.
        self.current_value = self.grid['x'+str(x_coordinate)].loc['y'+str(y_coordinate)]
        # try except block catches the errors where peripheral values at the edges of the dataframe cannot index out of range values
        try:
            north_west = self.grid['x'+str(x_coordinate-1)].loc['y'+str(y_coordinate-1)]
        except:
            north_west = None
        try:
            north = self.grid['x'+str(x_coordinate)].loc['y'+str(y_coordinate-1)]
        except:
            north = None
        try:    
            north_east = self.grid['x'+str(x_coordinate+1)].loc['y'+str(y_coordinate-1)]
        except:
            north_east = None
        try:
            west = self.grid['x'+str(x_coordinate-1)].loc['y'+str(y_coordinate)]
        except:
            west = None
        try:
            east = self.grid['x'+str(x_coordinate+1)].loc['y'+str(y_coordinate)]
        except:
            east = None
        try:
            south_west = self.grid['x'+str(x_coordinate-1)].loc['y'+str(y_coordinate+1)]
        except:
            south_west = None
        try:
            south = self.grid['x'+str(x_coordinate)].loc['y'+str(y_coordinate+1)]
        except:
            south = None
        try:
            south_east = self.grid['x'+str(x_coordinate+1)].loc['y'+str(y_coordinate+1)]
        except:
            south_east = None
        list_of_surrounding_values = [north_west, north, north_east, west, east, south_west, south, south_east]
        list_of_non_none_surrounding_values = []
        for i in list_of_surrounding_values:
            if i != None:
                list_of_non_none_surrounding_values.append(i)
            else:
                pass
        sum_of_surrounding_numbers = sum(list_of_non_none_surrounding_values)
        if self.current_value == 0:
            if sum_of_surrounding_numbers in (3, 6): #if its red and sum of numbers around it is 3 or 6 then change the value to 1 in the proposed new grid at the end of the iteration (generation) otherwise stay the same colour
                self.grid_to_be['x'+str(x_coordinate)].loc['y'+str(y_coordinate)] = 1
            else:
                pass
        elif self.current_value == 1:
            if sum_of_surrounding_numbers in (0, 1, 4, 5, 7, 8): #if its green and sum of numbers around it is 0, 1, 4, 5, 7 or 8 then change the value to 0 in the proposed new grid at the end of the iteration (generation) otherwise stay the same colour
                self.grid_to_be['x'+str(x_coordinate)].loc['y'+str(y_coordinate)] = 0
            else:
                pass
        else:
            pass
        self.current_value_to_be = self.grid_to_be['x'+str(x_coordinate)].loc['y'+str(y_coordinate)]
        
    def run_iteration(self):
        if isinstance(self.grid, pd.DataFrame):
            # If the grid variable is of type DataFrame then create a copy of the dataframe which serves as the new state in the next generation.
            # Goes through each value in the dataframe and applies the rules in the rules_for_colour_change() method for every value.
            self.grid_to_be = self.grid
            for col in self.grid:
                x_coordinate = int(list(col)[1])
                for row in range(len(self.grid[col])):
                    y_coordinate = row+1
                    if (x_coordinate == self.chosen_x) & (y_coordinate == self.chosen_y):
                        # If the value having its rules applied matches the coordinates of the chosen cell then adjust the counter depending on whether it is green (1) or not (0).
                        self.rules_for_colour_change(x_coordinate, y_coordinate) 
                        self.number_of_times_chosen_cell_is_green += self.current_value_to_be
                    else:
                        self.rules_for_colour_change(x_coordinate, y_coordinate) 
            # The new dataframe with all the rules applied to it becomes the current dataframe at the end of the iteration
            self.grid = self.grid_to_be
            self.number_of_generations += 1
            print('------------------------------------------------------------------------\n')
            print(f'Generation {self.number_of_generations}')
            print(f'So far your coordinates ({self.chosen_x},{self.chosen_y}) were green {self.number_of_times_chosen_cell_is_green} time(s)')
            display(self.grid) # please comment out if you wish to hide the progression of board states
            self.visualise_grid()  # please comment out if you wish to hide the progression of board states
        else:
            return
                            
    
    def play_game(self, x=20, y=20, chosen_x=1, chosen_y=1, N=1):
        self.chosen_x = chosen_x
        self.chosen_y = chosen_y
        self.N = N
        if (chosen_x < 1) | (chosen_x > x) | (chosen_y < 1) | (chosen_y > y):
            # Print messages if the user defines a cell outside of the bounds of the defined grid dimensions
            print('Please choose a cell within the bounds of the defined grid\n')
            print(f'Grid width: {x}')
            print(f'Grid height: {y}')
            print(f'Chosen x coodrinate: {chosen_x}')
            print(f'Chosen y coordinate: {chosen_y}')
        elif N < 0:
            print('Please use a positive integer to choose a generation (N)')       
        else:
            self.initialise_grid(x, y)
            self.visualise_grid()
            for number_of_generations in range(N): 
                self.run_iteration()
            if self.number_of_generations > 0:
                print(f'Total number of generations including Generation Zero: {self.number_of_generations+1}')
                print(f'Number of generations your chosen coordinates ({self.chosen_x},{self.chosen_y}) were green: {self.number_of_times_chosen_cell_is_green}')
            else:
                pass

if __name__ == '__main__':
    game_instance = Green_VS_Red()
    # The play_game() method takes arguments in the form (width of grid dimension, height of grid dimension, x coordinate of chosen cell, y coordinate of chosen cell, number of generations in which to check)
    # Please enter new values to test functionality
    # Please be aware the board states will be printed into the console
    # If you wish to hide the board states except Generation Zero please see the run_iteration() method on how to do that
    game_instance.play_game(20, 20, 5, 4, 7)
```
